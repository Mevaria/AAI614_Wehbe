## Random Walk

```python
from urllib.request import urlopen, Request
from bs4 import BeautifulSoup
import datetime
import random
import re

random.seed(datetime.datetime.now().strftime('%s'))
def getLinks(articleUrl):
    # Ensure there's a slash between the base URL and articleUrl
    if not articleUrl.startswith('/'):
        articleUrl = '/' + articleUrl
    req = Request(f'http://en.wikipedia.org{articleUrl}', headers={'User-Agent': 'Mozilla/5.0'})
    html = urlopen(req)
    bs = BeautifulSoup(html, 'html.parser')
    return bs.find('div', {'id':'bodyContent'}).find_all('a', href=re.compile('^(/wiki/)((?!:).)*$'))

count = 0
max_pages = 50
links = getLinks('/wiki/Kevin_Bacon')

while len(links) > 0 and count < max_pages:
    newArticle = links[random.randint(0, len(links)-1)].attrs['href']
    print(newArticle)
    links = getLinks(newArticle)
    count+=1
```

## Recursively crawling an entire site

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import re

pages = set()
def getLinks(pageUrl, max_pages=50):
    if len(pages) >= max_pages  or pageUrl in pages:
        return
    pages.add(pageUrl)
    print(pageUrl)

    try:
        req = Request(f'https://en.wikipedia.org{pageUrl}', headers={'User-Agent': 'Mozilla/5.0'})
        html = urlopen(req)

    except Exception as e:
        print(f"skip {pageUrl} -> {e}")
        return

    bs = BeautifulSoup(html, 'html.parser')

    for link in bs.find_all('a', href=re.compile('^(/wiki/)')):
        href = link.get("href")
        if ":" in href or "#" in href or "?" in href:
            continue

        if len(pages) >= max_pages:
            return
        getLinks(href, max_pages)

getLinks("/wiki/Wikipedia")
```

## Collecting Data Across an Entire Site

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import re

pages = set()
def getLinks(pageUrl, max_pages = 20):
    if len(pages) >= max_pages:
        return
    pages.add(pageUrl)
    try:
        url = f'https://en.wikipedia.org{pageUrl}'
        req = Request(url, headers={'User-Agent': 'Mozilla/5.0'})
        html = urlopen(req)
    except Exception as e:
        print(f"skip {url} -> {e}")
        return

    bs = BeautifulSoup(html, 'html.parser')
    try:
        title = bs.select_one("#firstHeading") or bs.find("h1")
        print(title.get_text(strip=True) if title else "(no title)")
        #mw-parser-output
        for p in bs.select('#bodyContent p'):
            txt = p.get_text(strip=True)
            if txt:
                first_para_text = txt
                break
        print(first_para_text if first_para_text else "(no body paragraph)")
        print('-'*20)
    except AttributeError:
        print('This page is missing something! Continuing.')

    for link in bs.find_all('a', href=re.compile('^(/wiki/)')):
        href = link.get('href')
        if ':' in href or '#' in href or '?' in href:
            continue
        if 'href' in link.attrs:
            if link.attrs['href'] not in pages:
                #We have encountered a new page
                newPage = link.attrs['href']
                if len(pages) >= max_pages:
                    return
                getLinks(newPage, max_pages)

getLinks('/wiki/General-purpose_programming_language')
```

## Crawling across the Internet

```python
from urllib.request import urlopen, Request
from urllib.parse import urlparse
from bs4 import BeautifulSoup
import random

def getInternalLinks(bs, url):
    netloc = urlparse(url).netloc
    scheme = urlparse(url).scheme
    internalLinks = set()
    for link in bs.find_all('a'):
        href = link.get('href')
        if not href:
            continue
        parsed = urlparse(href)
        # only keep http(s)
        if parsed.scheme not in ('', 'http', 'https'):
            continue
        if parsed.netloc == '':
            internalLinks.add(f'{scheme}://{netloc}/{href.strip("/")}')
        elif parsed.netloc == netloc:
            internalLinks.add(href)
    return list(internalLinks)

def getExternalLinks(bs, url):
    netloc = urlparse(url).netloc
    externalLinks = set()
    for link in bs.find_all('a'):
        href = link.get('href')
        if not href:
            continue
        parsed = urlparse(href)
        # only keep http(s)
        if parsed.scheme not in ('http', 'https'):
            continue
        if parsed.netloc and parsed.netloc != netloc:
            externalLinks.add(href)
    return list(externalLinks)

def getRandomExternalLink(startingPage):
    req = Request(startingPage, headers={
        'User-Agent': 'Mozilla/5.0',
        'Referer': 'https://www.google.com'
    })
    bs = BeautifulSoup(urlopen(req, timeout=15), 'html.parser')

    externalLinks = getExternalLinks(bs, startingPage)
    if not externalLinks:
        print('No external links, looking around the site for one')
        internalLinks = getInternalLinks(bs, startingPage)
        if not internalLinks:   # nothing left to try
            return None
        return getRandomExternalLink(random.choice(internalLinks))
    return random.choice(externalLinks)

def followExternalOnly(startingSite):
    externalLink = getRandomExternalLink(startingSite)
    if not externalLink:
        print('Gave up (no links found).')
        return
    print(f'Random external link is: {externalLink}')
    followExternalOnly(externalLink)

followExternalOnly('https://www.oreilly.com/')
```

## Collect all External Links from a Site

```python
from urllib.request import urlopen, Request
from bs4 import BeautifulSoup

# Collects a list of all external URLs found on the site
allExtLinks = []
allIntLinks = []
max_pages = 50

def getAllExternalLinks(url):
    if len(allIntLinks) + len(allExtLinks) >= max_pages:  # stop condition
        return

    try:
        req = Request(url, headers={'User-Agent': 'Mozilla/5.0'})
        bs = BeautifulSoup(urlopen(req, timeout=15), 'html.parser')
    except Exception as e:
        print(f"skip {url} -> {e}") # avoid crashing on 403/404/etc.
        return

    internalLinks = getInternalLinks(bs, url)
    externalLinks = getExternalLinks(bs, url)

    for link in externalLinks:
        if len(allIntLinks) + len(allExtLinks) >= max_pages:  # stop if max reached
            return
        if link not in allExtLinks:
            allExtLinks.append(link)
            print(link)

    for link in internalLinks:
        if len(allIntLinks) + len(allExtLinks) >= max_pages:  # stop if max reached
            return
        if link not in allIntLinks:
            allIntLinks.append(link)
            getAllExternalLinks(link)

allIntLinks.append('https://oreilly.com')
getAllExternalLinks('https://www.oreilly.com/')
```
